This is a notebook that will analyse the Housing Sales in Australia

```sql
 -- view top 5 columns

SELECT TOP 5 * 
FROM [House Property Sales Analysis].dbo.raw_sales
```

```sql
--ensure all the postcodes are 4 digits
SELECT len(postcode) as postcode_length
FROM [House Property Sales Analysis].dbo.raw_sales
```

```sql
-- Find out the date of the highest number of sales
SELECT
 TOP 1
 CAST(datesold as date) as Date_With_Highest_Sales
FROM [House Property Sales Analysis].dbo.raw_sales
GROUP BY datesold
ORDER BY MAX(price) DESC
```

```sql
-- Finding out the Postcode with the highest average price per sale

SELECT
 TOP 1
 postcode as Postcode_With_Highest_Sales
FROM [House Property Sales Analysis].dbo.raw_sales
GROUP BY postcode
ORDER BY MAX(price) DESC
```

```sql
-- Year with the lowest Sales
SELECT year(datesold) as lowest_sales_year
FROM [House Property Sales Analysis].[dbo].[raw_sales]
GROUP BY datesold
HAVING MIN(price) = (select MIN(price) FROM [House Property Sales Analysis].dbo.raw_sales)
```

```sql
-- Top six postcodes by years price
with CTE as (
    SELECT postcode, year(datesold) as years, price,
    DENSE_RANK() OVER(PARTITION BY year(datesold)  ORDER BY price desc ) as rank 
    FROM [House Property Sales Analysis].dbo.raw_sales
    )

SELECT TOP 6
    postcode, years, price
FROM CTE
WHERE rank <= 6
```
